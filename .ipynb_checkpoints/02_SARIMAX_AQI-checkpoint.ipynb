# AQI Forecasting — SARIMAX Model
**Dataset:** AnandVihar_AQI_data.csv  
**Target:** Daily mean AQI  
**Model:** SARIMAX(1,1,1)(1,1,1)[7] — captures weekly seasonality

SARIMAX extends ARIMA with a seasonal component and optional exogenous regressors.
Here we use pollutant sub-indices (PM2.5, PM10, NO₂, SO₂) as exogenous features.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.metrics import mean_squared_error, mean_absolute_error

print('Libraries loaded successfully.')
```

## 1. Load & Preprocess Data

```python
df = pd.read_csv('AnandVihar_AQI_data.csv', parse_dates=['Timestamp'])
df = df.set_index('Timestamp').sort_index()

# Interpolate missing values
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce').interpolate(method='time')

# Daily aggregation
daily = df.resample('D').mean().dropna(subset=['AQI'])

print(f'Daily records: {len(daily)}')
print(daily[['PM25_sub','PM10_sub','NO2_sub','AQI']].head())
```

## 2. Seasonal Decomposition

```python
decomp = seasonal_decompose(daily['AQI'], model='additive', period=7)
fig = decomp.plot()
fig.set_size_inches(14, 8)
fig.suptitle('Seasonal Decomposition of Daily AQI (period=7)', fontsize=13)
plt.tight_layout()
plt.savefig('sarimax_decomposition.png', dpi=150)
plt.show()
print('Weekly seasonal pattern clearly visible.')
```

## 3. Train/Test Split (with Exogenous Features)

```python
# Exogenous regressors: pollutant sub-indices
exog_cols = ['PM25_sub', 'PM10_sub', 'NO2_sub', 'SO2_sub']

# Fill any remaining NaNs in exog columns with forward fill
daily[exog_cols] = daily[exog_cols].ffill().bfill()

split_date = '2025-07-01'
train_y = daily['AQI'][:split_date].iloc[:-1]
test_y  = daily['AQI'][split_date:]
train_X = daily[exog_cols][:split_date].iloc[:-1]
test_X  = daily[exog_cols][split_date:]

print(f'Train: {len(train_y)} | Test: {len(test_y)}')
```

## 4. Fit SARIMAX(1,1,1)(1,1,1)[7]

Order: (p=1, d=1, q=1) non-seasonal  
Seasonal: (P=1, D=1, Q=1, s=7) weekly

```python
model = SARIMAX(
    train_y,
    exog=train_X,
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 7),
    enforce_stationarity=False,
    enforce_invertibility=False
)
fit = model.fit(disp=False)
print(fit.summary())
```

## 5. Forecast & Evaluation

```python
forecast = fit.forecast(steps=len(test_y), exog=test_X)
forecast.index = test_y.index

rmse = np.sqrt(mean_squared_error(test_y, forecast))
mae  = mean_absolute_error(test_y, forecast)
mape = np.mean(np.abs((test_y.values - forecast.values) / test_y.values)) * 100

print(f'SARIMAX(1,1,1)(1,1,1)[7] Performance:')
print(f'  RMSE : {rmse:.2f}')
print(f'  MAE  : {mae:.2f}')
print(f'  MAPE : {mape:.2f}%')
```

```python
pred_obj = fit.get_forecast(steps=len(test_y), exog=test_X)
ci = pred_obj.conf_int()
ci.index = test_y.index

fig, ax = plt.subplots(figsize=(14, 5))
train_y[-180:].plot(ax=ax, label='Train (last 180d)', color='steelblue')
test_y.plot(ax=ax, label='Actual', color='black', linewidth=1.5)
forecast.plot(ax=ax, label='SARIMAX Forecast', color='darkorange', linestyle='--')
ax.fill_between(ci.index, ci.iloc[:, 0], ci.iloc[:, 1], alpha=0.2, color='darkorange', label='95% CI')
ax.set_title(f'SARIMAX(1,1,1)(1,1,1)[7] Forecast  |  RMSE={rmse:.1f}  MAE={mae:.1f}', fontsize=13)
ax.set_ylabel('AQI')
ax.legend()
plt.tight_layout()
plt.savefig('sarimax_forecast.png', dpi=150)
plt.show()
```

## 6. Residual Diagnostics

```python
fit.plot_diagnostics(figsize=(14, 8))
plt.suptitle('SARIMAX Residual Diagnostics', fontsize=13)
plt.tight_layout()
plt.savefig('sarimax_diagnostics.png', dpi=150)
plt.show()
```

## 7. AIC/BIC Comparison (Model Selection)

If you want to try alternative orders:

```python
print(f'Selected model AIC : {fit.aic:.2f}')
print(f'Selected model BIC : {fit.bic:.2f}')
print()
print('Lower AIC/BIC = better model fit vs complexity trade-off.')
```

## 8. Summary

| Metric | Value |
|--------|-------|

| MAE | 151.35 |
| MAPE | ~62% |

**Key Takeaways:**
- SARIMAX explicitly models weekly seasonality (s=7).
- Exogenous pollutant features provide additional predictive signal.
- Despite richer structure, test-set RMSE is comparable to ARIMA.
- This suggests the test period (Jul–Dec 2025) has different distributional characteristics than training.
